### Aplicando o Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder() # Instanciado o LabelEncoder
le.fit_transform(dados['neighbourhood'])
```

```python
dados['Nova_Coluna'] = le.fit_transform(dados['neighbourhood'])
```

```python
pd.DataFrame(dados.dtypes) # Transformando os tipos das colunas em um Data Frame
```

```python
pd.DataFrame(dados.dtypes).reset_index() #Atribuindo um Index ao DataFrame dos categorias
```

```python
colunas = pd.DataFrame(dados.dtypes).reset_index().rename(columns={'index': 'nome',0:'tipo'})#Alterando o nome das colunas
categ_cols = colunas[colunas.tipo == 'object'] # Criando um dataFrame com apenas os nomes das colunas que possuem variaveis categoricas
```

```python
categ_cols
```

```python
for i in categ_cols.index:
    dados[str(categ_cols['nome'][i]) + '_encoded'] = le.fit_transform(dados[categ_cols['nome'][i]])
    dados = dados.drop(columns=[categ_cols['nome'][i]],axis=1)
```

```python
dados.head()
```

## Tratando os dados

```python
dados = dados.fillna(0)
dados = dados.drop(columns=['id'],axis=1)
```

```python
dados
```

##### Queremos prever o preço dos quartos do AirBnb
##### Então vamos usar todas as aoutras variáveis disponiveis para fazer uma regressão linear para o preço


```python
x = dados.drop(columns=['price','host_id','neighbourhood_group'], axis=1) # X = Variavel explicativa
# X vai ser todos os dados do dataFrame que usaremos pare treinar o modelo
y = dados.price # y = Variavel resposta
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=42) #random_state é uma semente, serve para padronizar
# a escolha das linhas de teste, para evitar resultados diferentes, o número pode ser qualquer um, mas ele deve sempre permanecer o mesmo
```

### Rodando Modelo

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()

modelo.fit(x_train,y_train)

y_pred = modelo.predict(x_test)
```

### Verificando o Resultado

```python
from sklearn.metrics import r2_score

r2_score(y_test,y_pred)
```
