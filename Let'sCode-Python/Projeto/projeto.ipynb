```python
def get_datasets(y,labels):
    if type(y[0]) == list:
        datasets = []
        for i in range(len(y)):
            datasets.append({
                'label':labels[i],
                'data':y[i]
            })
        return datasets
    else:
        return[
            {
                'label':labels[0],
                'data':y
            }
        ]
    
```

```python
def set_title(title=''):
    if title != '':
        display = 'true'
    else:
        display = 'false'

    return {
        'title':title,
        'display': display
    }
```

```python
def create_chart(x,y,labels,kind='bar',title=''):
    datasets = get_datasets(y,labels)
    options = set_title(title)
    
    chart = {
        'type': kind,
        'data':{
            'labels':x,
            'datasets':datasets  
        },
        'options':options
          
    }
    return chart
```

```python
def get_api_chart(chart):
    url_base = 'https://quickchart.io/chart'
    resp = r.get(f'{url_base}?c={str(chart)}')
    return resp.content
```

```python
# parametro 'wb' é para escrever arquivos binários 
def save_image(path,content):
    with open(path,'wb') as image:
        image.write(content)
```

```python
from PIL import Image
from IPython.display import display
```

```python
def display_image(path):
    img_pil = Image.open(path)
    display(img_pil)
```

Para o grafico que faremos é um plot de barras em grupos mostrando a evolução dos casos confirmados e o de recuperados, teremos dois dados diferentes para Y um dados de confirmados e de recuperados com base no DataSet que teremos com o consumo da API

```python
y_data_1 = []
for obs in final_data[1::10]:
    y_data_1.append(obs[CONFIRMADOS])
 
y_data_2 = []   
for obs in final_data[1::10]:
    y_data_2.append(obs[RECUPERADOS])
    
labels = ['Confirmados','Recuperados']

x = []
for obs in final_data[1::10]:
   x.append(obs[DATA].strftime('%d/%m/%Y')) #funcao strftime transforma uma variavel dateTime em uma variavel String
   
chart = create_chart(x,[y_data_1,y_data_2],labels,title='Gráfico confirmados VS recuperados')
chart_content = get_api_chart(chart)
save_image('Meu-primeiro-grafico.png',chart_content)
display_image('Meu-primeiro-grafico.png')
```

```python
def get_api_qrcode(link):
    text = quote(link) #parsing do link para URL
    url_base = 'https://quickchart.io/qr'
    resp = r.get(f'{url_base}?text={text}')
    return resp.content
    
```
